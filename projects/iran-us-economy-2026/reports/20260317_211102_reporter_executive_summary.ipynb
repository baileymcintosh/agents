## Historical Precedents & Lessons

The following table, drawn from Capital Street FX, Investopedia, Equirus, and the Baker Institute, places the 2026 shock in context:

| Crisis | Supply Loss | Price Change | Duration | Key Lesson |
|---|---|---|---|---|
| **1973 Arab Embargo** | ~7.5–8.6% of global supply | +300% ($3→$12) | 6 months | Political decision; reversed politically. Created IEA. |
| **1979 Iranian Revolution** | ~6% of global supply | +149% ($15→$39) | 12 months | Revolution = unpredictable duration; supply gap took 2–3 years to fill. |
| **1990 Gulf War** | ~8% of global supply | +141% ($17→$41) | 9 months | Swift military resolution → swift price retracement. |
| **2026 Iran War (current)** | Potentially 15–20% via Hormuz | +47% in 30 days; potentially +70–120% if sustained | Unknown | Largest potential supply disruption in modern history by magnitude. |

**Key parallels and divergences**, as synthesised by Capital Street FX (March 2026):
- **Like 1973:** Military conflict in the Middle East; supply shock intersecting with existing inflation concern; physical threat to oil transit infrastructure.
- **Unlike 1973:** The U.S. is now the world's largest oil producer (not a net importer); the IEA strategic reserve system holds **1.2 billion barrels of public stocks** (vs. zero in 1973) and has already activated its largest-ever coordinated release of **400 million barrels** on March 11 (LSE Business Review).
- **Like 1990:** Non-OPEC buffers exist; conflict involves a defined aggressor; futures market pricing suggests temporary, not permanent, disruption.
- **Unlike 1990:** The scale of potential Hormuz disruption — 20 mb/d — is **2.5x larger** than any prior crisis, per Investing Angles analysis. And this time, the Red Sea is simultaneously blocked.

The **1979 precedent** is the most sobering cautionary tale: a conflict involving Iran that was expected to resolve quickly instead lasted over a year, defied diplomatic forecasting, and contributed to stagflation that took the Volcker Fed's aggressive rate hikes to finally break.
