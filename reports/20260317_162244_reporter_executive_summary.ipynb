## Historical Precedents & Lessons

Five historical cases illuminate different dimensions of this conflict, each carrying a specific strategic lesson:

**1. The 1981 Osirak Strike (Iraq):** Israel destroyed Iraq's Osirak nuclear reactor in a single airstrike. Result: Saddam Hussein **accelerated and deepened** his covert nuclear program, which by 1990 was far more advanced than it had been in 1981. **Lesson:** Military strikes on nuclear programs that are disruptive but not eliminative tend to harden the target regime's determination. The Osirak effect appears to be operating in reverse on Iran — the June 2025 Midnight Hammer strikes intensified Iran's nuclear concealment activities rather than ending them, directly triggering the 2026 escalation. Both Builder reports and the Arms Control Association explicitly invoke this parallel.

**2. The 2003 Iraq Invasion:** The closest modern analogue in scale. Key lesson: air campaigns and decapitation strikes that achieve regime change rarely produce the political outcome imagined. The regime fell in 3 weeks; the insurgency lasted a decade. Operation Epic Fury has already struck more targets in 17 days than Operation Shock and Awe in its entire opening phase — but "regime change" via air power alone has no successful precedent in modern military history. Trump has not announced ground operations. Without a post-war political plan (The Guardian reported explicitly that there is "no plan"), the endgame is dangerously unclear.

**3. The 1973 Arab Oil Embargo:** The closest energy market analogue. The OPEC oil embargo quadrupled prices and triggered a global recession. The current Hormuz closure is potentially **more severe** — it affects not just oil but LNG, aluminum, fertilizer, and container trade simultaneously, and it cannot be resolved by a political agreement among oil-producing states. It requires either a military solution (reopening by force) or an Iranian decision to lift it voluntarily.

**4. The Iran-Iraq War (1980–1988):** Iran's most relevant strategic precedent for how it absorbs punishment. Over eight years of devastating conflict with 500,000+ casualties, Iran's regime demonstrated extraordinary resilience and tolerance for sacrifice. The "Tanker War" of 1984–1988, when both sides attacked each other's oil infrastructure and shipping, provides the most direct analogue for current Hormuz operations. Iran did not capitulate. **Lesson:** Air campaigns alone have never forced the Islamic Republic to surrender a strategic position. The regime has historically treated external military pressure as a source of domestic legitimacy.

**5. The Soleimani Assassination (January 2020):** The US targeted killing of IRGC Quds Force commander Qasem Soleimani. Iran retaliated with missiles at Al Asad airbase in Iraq, then de-escalated. That experience is being invoked by those who expected rapid Iranian capitulation in 2026. **The critical difference:** The Soleimani killing was a single targeted strike against one individual, not a massive air campaign that killed the Supreme Leader, his family, 40+ senior officials, degraded Iran's entire military, and closed the Strait of Hormuz. The scale difference is not incremental — it is categorical. Iran's response incentives in 2026 are incomparably higher than in 2020.

**Historical consensus across all Builder reports:** Military strikes on nuclear programs delay but rarely end nuclear ambitions, and frequently accelerate them through underground reconstitution. Decapitation strikes are not reliably war-ending. Economic coercion via chokepoints (1973) can trigger global recessions. The regime that has survived 46 years of sanctions, wars, and assassinations will not be broken quickly by air power alone.
